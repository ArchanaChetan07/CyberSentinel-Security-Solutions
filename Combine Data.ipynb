### Combine Data Tables into a Single Table

- The goal is to combine the NB15 into a single table. Then combine the CIC tables into one. Finally, combine the NB15 and CIC tables.

- NOTE: Columns in the NB15 dataset will need to be normalized to match the classification of the CIC dataset. This includes updating the Duration, matching the Protocol type, and changing the Label values

#### Combine all four NB15 datasets

```python
df1 = nb15_1[['Source IP', 'Source Port', 'Destination IP', 'Destination Port', 'Protocol', 'Flow Duration', 'Total Fwd Packets', 'Total Backward Packets', 'Fwd Packet Length Mean', 'Bwd Packet Length Mean', 'Label']]
df2 = nb15_2[['Source IP', 'Source Port', 'Destination IP', 'Destination Port', 'Protocol', 'Flow Duration', 'Total Fwd Packets', 'Total Backward Packets', 'Fwd Packet Length Mean', 'Bwd Packet Length Mean', 'Label']]
df3 = nb15_3[['Source IP', 'Source Port', 'Destination IP', 'Destination Port', 'Protocol', 'Flow Duration', 'Total Fwd Packets', 'Total Backward Packets', 'Fwd Packet Length Mean', 'Bwd Packet Length Mean', 'Label']]
df4 = nb15_4[['Source IP', 'Source Port', 'Destination IP', 'Destination Port', 'Protocol', 'Flow Duration', 'Total Fwd Packets', 'Total Backward Packets', 'Fwd Packet Length Mean', 'Bwd Packet Length Mean', 'Label']]

comb_nb15 = pd.concat([df1, df2, df3, df4], ignore_index = True)

comb_nb15.head()
```

#### Combine the CIC datasets

```python
df5 = cic161718[['Src IP', 'Src Port', 'Dst IP', 'Dst Port', 'Protocol', 'Flow Duration', 'Tot Fwd Pkts', 'Tot Bwd Pkts', 'Fwd Pkt Len Mean', 'Bwd Pkt Len Mean', 'Label']]
df6 = cic2019[[' Source IP', ' Source Port', ' Destination IP', ' Destination Port', ' Protocol', ' Flow Duration', ' Total Fwd Packets', ' Total Backward Packets', ' Fwd Packet Length Mean', ' Bwd Packet Length Mean', ' Label']]

df5 = df5.rename(columns = {'Src IP':'Source IP', 'Src Port':'Source Port', 'Dst IP':'Destination IP', 'Dst Port':'Destination Port', 'Protocol':'Protocol', 'Flow Duration':'Flow Duration', 'Tot Fwd Pkts':'Total Fwd Packets', 'Tot Bwd Pkts':'Total Backward Packets', 'Fwd Pkt Len Mean':'Fwd Packet Length Mean', 'Bwd Pkt Len Mean':'Bwd Packet Length Mean', 'Label':'Label'})
df6 = df6.rename(columns = {' Source IP':'Source IP', ' Source Port':'Source Port', ' Destination IP':'Destination IP', ' Destination Port':'Destination Port', ' Protocol':'Protocol', ' Flow Duration':'Flow Duration', ' Total Fwd Packets':'Total Fwd Packets', ' Total Backward Packets':'Total Backward Packets', ' Fwd Packet Length Mean':'Fwd Packet Length Mean', ' Bwd Packet Length Mean':'Bwd Packet Length Mean', ' Label':'Label'})

comb_cic = pd.concat([df5, df6], ignore_index = True)

comb_cic.head()
```

#### Normalize the data between both combined datasets

```python
comb_nb15['Flow Duration'] = comb_nb15['Flow Duration'] * 1000
comb_nb15['Label'] = comb_nb15['Label'].replace({0:'benign', 1:'ddos'})

comb_cic['Protocol'] = comb_cic['Protocol'].replace({0:'Unknown', 6:'tcp', 17:'udp'})
comb_cic['Label'] = comb_cic['Label'].replace({'DDOS':'ddos', 'DDoS':'ddos', 'BENIGN':'benign', 'Benign':'benign'})
```

#### Combine everything and output it to a .csv file

```python
final_df = pd.concat([comb_cic, comb_nb15], ignore_index = True)

final_df.head()
```

```python
final_df.to_csv('MSADS 508 Final Project.csv', index = False)
```
